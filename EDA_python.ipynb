```python
import pandas
import numpy 
import matplotlib.pyplot as plt
import seaborn
```

```python
ipl=pandas.read_csv(".\\DONNEES\\matches.csv", sep=",",
                   decimal=".", encoding="utf-8")
```

```python
ipl.head() ## afficher les 5 premières lignes
```

```python
## afficher les dimensions d'ipl :
ipl.shape
```

Ensuite, il faut décrire les variables :qu'est-ce que leur nom dignifie ? Quelle est leur unité ? Donner des exemples de lecture.

```python
## obtenir la fréquence des matchs gagnés par équipe :
ipl["player_of_match"].value_counts()
```

```python
## obtenir les 10 premiers équipes qui gagnent les matchs :
ipl["player_of_match"].value_counts()[0:10]
```

```python
list(ipl["player_of_match"].value_counts()[0:5].keys())
```

```python
## barplot pour les 5 premiers équipes gagnantes :
plt.figure(figsize=(8,5))
plt.bar(list(ipl["player_of_match"].value_counts()[0:5].keys()),
        list(ipl["player_of_match"].value_counts()[0:5]),
        color="g"
       )
         
plt.show()
```

```python
ipl["result"].value_counts()
```

```python
## compter le nb de lancés gagnants :
ipl["toss_winner"].value_counts()
```

```python
## les enregistrements où une équipe a gagné la première battue
batting_first=ipl[ ipl["win_by_runs"] !=0 ]
batting_first.head()
```

```python
## faire un histogram :
plt.figure(figsize=(8,5))
plt.hist(batting_first["win_by_runs"],
        edgecolor="white")
plt.title("Distribution de runs", fontsize=30)
plt.xlabel("runs", fontsize=15)
plt.show()
```

Il y a environ 80 équipes qui ont gagné entre 15 et 25 matches

```python
## ensuite on veut avoir la distribution des gagnants :
batting_first["winner"].value_counts()
```

```python
## représenter les 3 premières équipes :
## faire un histogram :
plt.figure(figsize=(8,5))
plt.bar(list(batting_first["winner"].value_counts()[0:3].keys()),
        list(batting_first["winner"].value_counts()[0:3]),
        color=["blue", "yellow", "green"])

plt.show()
```

```python
## faire un piechart :
plt.figure(figsize=(7,7))
plt.pie(
    list(batting_first["winner"].value_counts()),
    labels=list(batting_first["winner"].value_counts().keys()),
    autopct="%0.1f%%"
)
plt.show()
```

```python
## quelles sont les équipes qui ont gagné après avoir été battue :
batting_second = ipl[ ipl["win_by_wickets"] != 0 ]
batting_second.head()
```

```python
## faire un histogram :
plt.figure(figsize=(8,5))
plt.hist(batting_second["win_by_wickets"],
         bins=30,
         edgecolor="white")
plt.title("Distribution de wickets", fontsize=30)
plt.xlabel("wickets", fontsize=15)
plt.show()
```

```python
batting_second["winner"].value_counts()
```

```python
## faire un barplot :
plt.figure(figsize=(8,5))
plt.bar(list(batting_second["winner"].value_counts()[0:3].keys()),
        list(batting_second["winner"].value_counts()[0:3]),
        color=[ "white", "grey", "black"],
        edgecolor="black")
plt.show()
```

```python
## faire un piechart :
plt.figure(figsize=(7,7))
plt.pie(
    list(batting_second["winner"].value_counts()),
    labels=list(batting_second["winner"].value_counts().keys()),
    autopct="%0.1f%%"
)
plt.show()
```
