## 2/ Analyse bivariée <a class="anchor" id="p2"></a>
### 2.1/ Test de corrélation de Bravais Pearson : r <a class="anchor" id="rbrav"></a>

<div style="test-align:justify">
    <p><span style="font-weight:bold;">Définitions</span>. Le R de Bravais Pearson permet de mesurer l'association entre
        deux <span style="color:red;">variables quantitatives continues</span>. Il peut également être utilisé dans le cas d'une analyse multivariée. On teste l'indépendance entre deux variables quantitatives continues ou plus, avec une variable explicative (ou
        variable indépendante) et une variable à expliquer (ou variable dépendante). Dans le cas du ML, ce test est
        intéressant pour exclure les variables explicatives qui n'apportent pas d'information. Le R de Bravais-Pearson est
        caractérisé par l'intensité d'une relation (<span style="font-weight:bold; color:red;">coefficient de corrélation</span>), la
        part totale qu'explique la variable explicative (<span style="font-weight:bold; color:red;">le coefficient de
        détermination</span>).</p>
    <p><span style="font-weight:bold;">Conditions d'application</span>. 
    <li>Le test d'indépendance du R de Bravais-Pearson est utilisé si, après avoir représenté graphiquement les deux variables sous la forme d'un nuage de points</li>, 
    on peut
        "simplifier" la relation entre ces deux variables par une droite. Enfin, il n'est pas nécessaire que les variables aient une distribution normale pour pouvoir exécuter le R de Bravais-Pearson.</p>
    <p><span style="font-weight:bold;">Interprétation des résultats</span>. Le R de Bravais-Pearson varie entre -1 (relation très intense négative) et 1 (relation très intense positive), et où 0 indique l'absence de relation. Cette relation est fiabilisée par le hasard : la p-value. Au seuil de 5% (si p > 5%), on peut dire qu'il n'y a pas indépendance, et que H0 peut être rejeté, tandis qu'on accepte H1 avec 5% de chance de faire une erreur de rejeter H0. Enfin le coefficient de détermination, appelé R², est la part expliquée par la relation. Ce test est sensible à la taille de l'échantillon. Plus l'échantillon est grand plus la p-value sera basse et le R mieux estimé.</p>
    <p>Afin de performer un tel test, nous avons récupéré un jeu de données sur le fromage, sur le <a href="https://github.com/asardell/statistique-python/blob/master/Dataset/fromage.txt">github de asardell</a> qui fourni également des scripts Python pour les calculs de coefficient de corrélation</p>
</div>

````python
import pandas ## manipuler des données
import seaborn ## pour la dataviz
from scipy.stats import pearsonr ## performer test statistique

f=pandas.read_csv(".\\DONNEES\\fromage.txt", sep="\t", decimal=".", encoding="utf-8", index_col=0)
f.head()

## On commence par projeter des paires de variables sous forme de nuage de points :
seaborn.pairplot(f.iloc[:,0:9])
````

````python
## Après avoir utilisé des scatterplots pour afficher de possibles relations linéaires,
##+ il est nécessaire d'étudier les corrélations possibles entre les variables :
seaborn.set(rc={'figure.figsize':(10,4)})

df_corr = f.corr() ## matrice de corrélation

ax = seaborn.heatmap(df_corr, xticklabels = df_corr.columns, 
                     yticklabels = df_corr.columns, cmap = 'coolwarm')
````

````python
## De cette matrice de corrélation, 
##+ on peut identifier des variables ayant des corrélations fortes et faibles, positives ou négatives.
## On doit maintenant calculer leur fiabilité à l'aide la p-value :
pearsonr(f.lipides, f.magnesium)
````

On peut dire qu'au risque de 5%, on peut rejeter H0 et accepter H1 car la p-value (pvalue) est inférieur au seuil de 5%. Et aussi, R² (statistic) est de 0.69.

### 2.3/ Analyse de variance 1 facteur <a class="anchor" id="anova_oneway"></a>

**Définition**. Une ANOVA (analyse de variance) permet de mesurer l'indépendance entre une variable qualitative et une variable quantitative. Dans un tel test, on cherche à savoir si les moyennes des groupes sont significativement différentes. On va donc poser deux hypothèses : H0 : les moyennes de chaque groupe sont égales, si la p-value > 5% : H1 : les moyennes de chaque groupe ne sont pas toutes égales si la p-value est inférieure à 5%.
**Conditions d'application**. On cherche à comparer deux moyennes ou plus, de groupes. Il faut donc que chaque groupe dispose de deux individus ou plus. Si jamais il n'y a qu'un seul individu, alors l'ANOVA ne peut être opérée, et il faudra retirer le groupe entier. Les résultats d'un tel test sont valides, si :

- l'indépendance entre les échantillons de chaque groupe est validée (tout dépend du contexte),
- l'égalité des variance est assurée. On utilise pour ça un boxplot OU un test de Bartlett.
- Les résidus ont une distribution normale (égalité de la moyenne et de la médiane, distribution en cloche ou test de Shapiro-wilk).

**Remarque** : Dans la réalité, les trois conditions ne sont pas toujours toutes vérifiées en tous cas en sciences sociales. Il faut cependant vérifier et avoir le plus d'hypothèses de validités vérifiées (Dumolard, 1997).

[Le jeu de données d'entraînement](https://github.com/asardell/statistique-python/blob/master/Dataset/Hotdogs.csv)  
[Comment faire une ANOVA avec comparaison des facteurs](https://www.reneshbedre.com/blog/anova.html#google_vignette)

````python
import pandas 
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn
````

````python
df=pandas.read_csv(".\\DONNEES\\Hotdogs.csv", sep=";", encoding="utf-8")
df.head()
````

````python
df.Type.unique()
````

````python
plt.subplots(figsize=(20,4))
ax = seaborn.boxplot(x="Calories", y="Type", data=df)
````

On rappelle les hypothèses :
    
- H0 : les moyennes de chaque groupe sont égales si la p-avlue > 5%,
- H1 : les moyennes de chaque groupe ne sont pas toutes égales < 5%

````python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('Calories ~ Type', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
````

PR(>F) est la p-value. Ici on a une p-value bien inférieure au seuil de 5%.

## 4/ Analyse multivariée : modélisation <a class="anchor" id="p4"></a>

Afin de réaliser cette partie, on va utiliser le cours de modélisation avec R dispensé en master 1. Dans un premier temps, il est important d'étudier le principe de la modélisation, puis étudier la modélisation par la régression linéaire et les cas particulier de la droite des moindres carrés (ols). <span style="color:red; font-weight:bold;">Afin de ne pas répéter le cours de Feuillet, nous allons prendre quelques notes ici, pour justifier les étapes de calcul du modèle. **Qu'est-ce que la modélisation ?** Il existe différents types de modèles. En sciences sociales, les modèles sont des simplifications de la réalité, tenant compte de l'erreur issue de la mesure ou de phénomène non identifié. Ainsi, une équation de modèle prend la forme :
f(x) = axi + b + ei,
avec a le coefficient directeur, xi la valeur de l'individu sur lequel on utilise le modèle, b la constante et e l'erreur stochastique associée à l'individu. **Le e est un résidus qui est la différence entre les valeurs observées et les valeurs théoriques**. Ainsi, les modèles sont également appelés "modèle stochastique". La valeur et le signe de a renseigne sur la force de la liaison et son sens. On l'obtient en calculant la variation de Y sur X. Tandis que b est l'ordonnée à l'origine et est identifié quand X = 0.

### 4.1/ Régression linéaire : droite des moindres carrés <a class="anchor" id="regLin_ols"></a>

**Qu'est-ce que la régression linéaire ?** Une régression linéaire consiste en la modélisation de la covariance entre deux variables : Y à expliquer et X explicative. On dit que X est **indépendante**. On utilise les régressions pour :
- prédire : quelle est la valeur pour un individu dans le passé ou dans le futur ?
- comprendre : explorer la présence et la force de la relation entre les deux variables choisies.

La régression linéaire est donc une équation, qui va être représentée par une droite.

**Conditions d'application** :
- X et Y sont deux variables quantitatives continues.
- X peut être une variable qualitative.
- La relation entre X et Y doit sembler linéaire selon le nuage de points et selon le résultat du coefficient de corrélation.
- Définir deux hypothèses : H0 : indépendance entre X et Y. H1 : X et Y ne sont pas indépendants.

La droite de l'équation de régression linéaire est une droite passant par le nuage de points, et qui ajuste le mieux ce nuage de points. Autrement dit, c'est la droite qui minimise la distance entre tous les points. A partir de cette droite, il sera facile de lire graphiquement quelle serait la valeur de Y en fonction d'un X non mesuré.

Exemple : 


Source : https://towardsdatascience.com/linear-regression-using-python-b136c91bf0a2 

**Conditions de validité et fiabilisation** :
Bien que le modèle permette de faire de la prédiction ou de décrire des relations, il est nécessaire de <span style="color:red; font-weight:bold;">mesurer sa qualité</span>. Dans ce cadre, on utilise le R², qui permet d'obtenir le pourcentage de variance de Y expliqué par le modèle :



Ce R² varie entre 0 et 1 (donc entre 0 et 100%) si on tient compte uniquement du R² (et qu'on ne le passe pas en pourcentage).

Par la suite, il faut également estimer la <span style="color:red; font-weight:bold;">significativité de la relation entre X et Y</span>. A cette fin, on utilise les statistiques inférentielles. On rappelle que l'OLS permet d'avoir un estimateur des vraies valeurs de a et b. Dans ce cas, il faut être certain que a et b sont des valeurs significativement différentes de 0 (c'est-à-dire que leur vraie valeur n'est pas 0). Pour cela, on va utiliser la t-value :

- t-value : plus la t-value est proche de 0, plus on se rapproche d'un effet de X sur Y nul. Il faut que la t-value soit supérieure à 1,96 ou inférieure à 1,96 pour avoir 95% chance de rejeter H0 dans se tromper. Ensuite on choisi un seuil alpha (généralement 5%) auquel on va comparer la t-value estimée. Quand t > t_alpha, on peut rejeter H0.

- p-value : significativité du résultat. 

Enfin, plusieurs hypothèses doivent être validées pour s'assurer que les estimateurs sont les moins biaisés possibles. Ce sont les estimations BLUE. Ces estimations s'opérent sur les résidus. Les hypothèses stochastiques à valider :

- les résidus ont une distribution normale et une moyenne nulle,
- la variance est constante,
- indépendance des résidus entre eux

## 6/ Techniques de représentations de données <a class="anchor" id="p6"></a>

### Description des modules
#### Plotly
Plotly est un module Python qui va s'appuyer sur plotly.js pour fournir des charts interactifs. Il peut être utilisé dans les notebooks comme dans les applications Python. Pour utiliser plotly dans un jupyter notebook, il faut d'abord avoir installé : pandas, seaborn, numpy et matplotlib. Ensuite, dans le Notebook, il faut exécuter ```from plotly_offline import init_notebook_mode(connected=True)```.
Plotly est composé de plusieurs sous-modules :

- plotly.tools : module qui permet de mieux contrôler les charts.
- plotly.graphs_objs : crée les graphes, les figures, la donnée et les sorties (_layout_).
- plotly.plotly : un ensemble de fonction appelées pour utiliser les ressources de plotly server (on rappelle qu'on utilise une API).

Afin de pouvoir utiliser plotly sur des dataframe pandas, on va utiliser le module cufflinks.

#### Cufflinks
N'est pas obligatoire pour utiliser plotly seul. Utilisé pour connecter les dataframe de Pandas avec plotly. Après avoir importé le module, exécuter `cufflinks.go_offline()`.
