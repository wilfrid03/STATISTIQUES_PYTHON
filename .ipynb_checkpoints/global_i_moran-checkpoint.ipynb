<p style="text-align:justify;">"L'autocorrélation spatiale a donc trait au degré auquel la similarité des valeurs entre les observations d'un ensemble de données est liée à la similarité des emplacements de ces observations." Elle met en relation la valeur d'une variable avec ses autres emplacements.</p>

<h2 style="border:2px solid green;">Comprendre l'autocorrélation spatiale</h2>
<p style="text-align:justify; padding:1%;">Imaginons que l'autocorrélation spatiale n'existe pas. Autrement dit, l'emplacement d'une observation n'informe pas sur sa valeur. On parle alors de distribution spatialement aléatoire, puisqu'aucune structure spatiale émerge de la localisation des valeurs. Et alors ? Est-ce nécessaire d'avoir une telle relation ? Savoir qu'une valeur est liée à un espace permet de caractériser cet espace. Donc si une même valeur se trouve dans plusieurs espaces, alors elle caractérise ces espaces : on peut dire qu'ils se "ressemblent". L'autocorrélation est caractérisée par une échelle et un signe. Quand son signe est positif, alors on dit que les valeurs similaires ont tendance à être proches. Par exemple : les revenus. Sinon, quand le signe est négatif, alors on dit que les valeurs similaires ont tendance à s'éloigner : par exemple certains magasins, ou les hôpitaux, afin d'avoir la couverture spatiale la plus pertinente. Ensuite, l'échelle : l'autocorrélation peut être soit globale (informe sur la tendance générale du jeu de données), ou locale (permet de déterminer les clusters). Ici, on va se concentrer uniquement sur l'autocorrélation globale.</p>
<p style="text-align:justify; padding:1%;">"Nous explorerons ces concepts à l'aide d'un exemple appliqué, en <span style="font-weight:bold">interrogeant les données sur la présence, la nature et la force de l'autocorrélation spatiale globale</span>. Pour ce faire, <span style="font-weight:bold;">nous utiliserons un ensemble d'outils connus collectivement sous le nom d'analyse exploratoire des données spatiales (AEDS)</span>. Analogue à son équivalent non spatial (EDA), l'AEDS a été spécialement conçue à cette fin et place l'espace et l'emplacement relatif des observations dans un ensemble de données au premier plan de l'analyse. <span style="font-weight:bold">L'éventail des méthodes d'ESDA est large et va des approches les plus simples, comme les cartes choroplèthes (chapitre précédent), à des méthodologies plus avancées et plus robustes qui incluent l'inférence statistique et une reconnaissance explicite de l'arrangement géographique des données.</span> L'objectif de ce chapitre est de nous plonger dans ce dernier groupe."</p>

```python
## Les modules nécessaires :
# Graphics
import matplotlib.pyplot as plt
import seaborn
import libpysal 
## from pysal.viz import splot
## from splot.esda import plot_moran ## pour affichage du plot de Moran principalement

# Analysis
import geopandas
import pandas
from libpysal import weights
from numpy.random import seed

print(libpysal.__version__)
print(geopandas.__version__)
```

```python
## charger les données :
ref = pandas.read_csv("https://geographicdata.science/book/_downloads/3c5d077df052a9a21915c1a938536f99/brexit_vote.csv",
                       index_col="Area_Code")
ref.info()
```

```python
ref.head()
```

```python
lads = geopandas.read_file("https://geographicdata.science/book/_downloads/860cf1380fe6cb15c80f882fad461a04/local_authority_districts.geojson").set_index("lad16cd")

lads.head() ## retourne des MULTIPOLYGON
```

<p>On va se concentrer uniquement sur Pct_Leave qui est la proportion de votes en faveur de l'alternative Leave. On va utiliser la projection Mercator, préférée pour la cartographie (Folium, matplotlib et contextily).</p>

```python
db = (
    geopandas.GeoDataFrame(
        lads.join(ref[["Pct_Leave"]]), crs=lads.crs
    )
    .to_crs(epsg=3857)[
        ["objectid", "lad16nm", "Pct_Leave", "geometry"]
    ]
    .dropna()
)
db.info()
```

```python
## db = geopandas.read_file(".\\TRAITEMENT\\db.geojson", encoding="utf-8", crs=3857, geometry="geometry")     
```

<p>La couche db a fait l'objet de nombreuses tentatives de corrections. La reprojection en mètre était impossible, parce que les valeurs x,y,z et time étaient différents. Je n'ai pas trouvé le moyen de corriger automatiquement ce genre d'erreur. Du coup, après vaoir préparé les données conformément au manuel, la couche a été passée dans QGIS pour vérification géométrique et reprojection en 3857.</p>

<p>On réalise une carte choropleth pour avoir une première idée de la distribution des données et formuler des hypothèses.</p>

```python
db.Pct_Leave.plot(kind="hist")
```

```python
import contextily

f, ax = plt.subplots(1, figsize=(9, 9))
db.plot(
    column="Pct_Leave",
    cmap="viridis",
    scheme="quantiles",
    k=5,
    edgecolor="black",
    linewidth=0.5,
    alpha=0.75,
    legend=True,
    legend_kwds={"loc": 2},
    ax=ax,
)

contextily.add_basemap(
    ax=ax,
    crs=db.crs.to_string(),
    source=contextily.providers.CartoDB.Positron
)

plt.title("Quantiles des votants")
ax.set_axis_off()
```

<p>A ce stade, on peut déjà voir un genre de structure spatiale émerger. Seulement, elle est dépendante de la discrétisation que nous avons réalisé. Dans ce cas, nous devons vérifier s'il existe une réelle tendance au regroupement ou bien si c'est le fruit uniquement de la perception et de la discrétisation. C'est l'autocorrélation globale.</p>

<h3>Le décalage spatial</h3>

<p style="text-align:justify;">On a demandé à ChatGPT qu'est-ce qu'n lissage en statistique. Voici sa réponse "En statistiques, le lissage (ou "smoothing" en anglais) est une <span style="font-weight:bold;">technique utilisée pour atténuer les fluctuations aléatoires ou le bruit dans un ensemble de données</span>. Le but du lissage est de créer une représentation plus régulière des données, ce qui peut <span style="font-weight:bold;">rendre plus facile l'interprétation des tendances et des modèles qui sous-tendent les données</span>.</p>
<p style="text-align:justify;">Le lissage est souvent utilisé pour traiter des données temporelles, où les mesures sont effectuées à des intervalles réguliers dans le temps, comme les données météorologiques ou les données de marché financier. Dans ce contexte, le lissage peut être utilisé pour réduire les fluctuations aléatoires et mettre en évidence les tendances à plus long terme.</p>

<p style="text-align:justify;">Il existe différentes techniques de lissage, telles que le lissage exponentiel, le lissage de moyenne mobile et le lissage de régression. Chacune de ces techniques a ses avantages et ses inconvénients, et le choix de la technique de lissage appropriée dépendra du type de données et des objectifs de l'analyse."</p>

```python
# Generate W from the GeoDataFrame
## calcul du décalage spatial :
w = weights.KNN.from_dataframe(db, k=8)
# Row-standardization
w.transform = "R"
```

```python
db["Pct_Leave_lag"] = weights.spatial_lag.lag_spatial(
    w, db["Pct_Leave"]
)

## Afficher les résultats :
db.loc[["E08000012", "S12000019"], ["Pct_Leave", "Pct_Leave_lag"]]
```

<p style="border:2px solid red; padding:5px;"><span style="font-weight:bold;">Interprétation : </span>la première ligne (Liverpool), qui était reconnu comme "Remainer" parmi les plus "Leave" du nord de l'Angleterre. "En dehors de Londres et de l'Écosse, c'est l'un des rares endroits où moins de la majorité des électeurs ont voté pour le parti." La seconde ligne "représente Midlothian, en Écosse, où aucune autorité locale n'a voté en faveur de la sortie. Bien que Liverpool et Midlothian affichent un pourcentage similaire de population ayant voté en faveur de la sortie (42 % et 38 %, respectivement), la différence entre leurs décalages spatiaux reflète le contexte géographique plus large, qui est tout à fait différent"</p>

```python
f, axs = plt.subplots(1, 2, figsize=(12, 6))
ax1, ax2 = axs

db.plot(
    column="Pct_Leave",
    cmap="viridis",
    scheme="quantiles",
    k=5,
    edgecolor="white",
    linewidth=0.0,
    alpha=0.75,
    legend=True,
    ax=ax1,
)
ax1.set_axis_off()
ax1.set_title("% Leave")
contextily.add_basemap(
    ax1,
    crs=db.crs,
    source=contextily.providers.Stamen.TerrainBackground,
)

db.plot(
    column="Pct_Leave_lag",
    cmap="viridis",
    scheme="quantiles",
    k=5,
    edgecolor="white",
    linewidth=0.0,
    alpha=0.75,
    legend=True,
    ax=ax2,
)
ax2.set_axis_off()
ax2.set_title("% Leave - Spatial Lag")
contextily.add_basemap(
    ax2,
    crs=db.crs,
    source=contextily.providers.CartoDB.Positron,
)
plt.show()
```
