```python
## On va réaliser une régression linéaire sur un jeu de données fictifs : score à un examen et heures d'études
## on va tenter d'adapter un modèle de régression linéaire :

import pandas

## jeu de données fictifs :
df = pandas.DataFrame({'hours': [1, 2, 4, 5, 5, 6, 6, 7, 8, 10, 11, 11, 12, 12, 14],
                   'score': [64, 66, 76, 73, 74, 81, 83, 82, 80, 88, 84, 82, 91, 93, 89]})
      
## on commence par visualiser le jeu de données
## On doit s'assurer que visuellement il y a une tendance à la corrélation linéaire
## c'est un pré-requis pour réaliser une régression linéaire :
```

```python
## pour ça on va faire un scatter plot :
import matplotlib.pyplot as plt 

plt.scatter(df.hours, df.score)
plt.title("Heures d'études et score :")
plt.xlabel("Heures d'études")
plt.ylabel("Score obtenu")
plt.show()
```

<p style="padding:5px; border:1px red solid">Il y a une tendance à la linéarité : plus il y a d'heure de travail, plus le score est élevé. On peut ensuite utilise un boxplot pour voir s'il y a des outliers.</p>

```python
df.boxplot(column=['score'])
```

<p style="padding:5px; border:1px red solid">Il semblerait qu'il n'y ait pas d'outliers. On peut passer à la modélisation. On va utiliser le modèle OLS (droite des moindres carrés ordinaires).</p>

```python
import statsmodels.api as sm
## choisir la variable réponse (y) :
y = df['score']

## Choisir la variable exploratoire (x) :

#define response variable

#define explanatory variable
x = df[['hours']]

# Ajouter une constante :
x = sm.add_constant(x)

# fit linear regression model
model = sm.OLS(y, x).fit()

# view model summary
print(model.summary())
```

<p style="padding:5px; border:1px red solid">P > t = la p-value ; le R-squared = pourcentage de variation de y, expliqué par x (ici, x explique 83% de la variation). Ensuite, il faut créer un residual plot. L'une des hypothèses de la régression linéaire est que les résidus sont distirbués normalement et homoscédastique à chaque niveau de la variable explicative. Si ces hypothèses de sont pas respectées, le modèle n'est pas fiable. On peut vérifier visuellement ces hypothèses : </p>
<p style="padding:5px; border:1px red solid">"Graphique des valeurs résiduelles par rapport aux valeurs ajustées : Ce graphique est utile pour confirmer l'homoscédasticité. L'axe des x affiche les valeurs ajustées et l'axe des y affiche les résidus. Tant que les résidus semblent être distribués de manière aléatoire et uniforme dans le graphique autour de la valeur zéro, nous pouvons supposer que l'homoscédasticité n'est pas violée :"</p>

```python
#define figure size
fig = plt.figure(figsize=(12,8))

#produce residual plots
fig = sm.graphics.plot_regress_exog(model, 'hours', fig=fig)
```

<p style="padding:5px; border:1px red solid">Le graphique en haut à gauche nous indique comment se situent les valeurs fitted par rapport aux valeurs observées. On peut voir une certaine proximité pour certaines valeurs. Ensuite il faut vérifier la normalité des résidus. Pour ça, on peut utiliser un q-q plot. Si les résidus sont proches d'une droite à 45° du point do'irgine (0, 0), alors les résidus suivent une loi normale :</p>

```python
#define residuals
res = model.resid

#create Q-Q plot
fig = sm.qqplot(res, fit=True, line="45")
plt.show() 
```

<p style="padding:5px; border:1px red solid">Malgré un décalage, on peut dire que les résidus sont normaux. Le décalage est ici considéré comme impactant légèrement le résultat.</p>

<p style="padding:5px; border:1px red solid">Nous avons vérifié l'homoscédasticité des résidus ainsi que leur normalité. Notre modèle de régression linéaire simple est donc fiable.</p>
